```python
# Step 1: Import our Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
%matplotlib inline
# We use this to have graph displayed in a cell, just like the other results
```

```python
# Step 2: Load our Data
tree_data = pd.read_csv("nyc_trees.csv")
```

```python
# Step 3: Do you really know your dataset?
tree_data.head()
```

```python
tree_data.describe()
```

```python
# Step 4:
plt.hist("age", data=tree_data)
# Let's do some Work, work, work, work, work, work...
```

```python
# Plot 1: Histogram
plt.hist("age", data=tree_data, bins = 20)
```

```python
# Plot 2: Scatter Plot
plt.scatter(x="age", y="diameter", data=tree_data, color="#899481")
```

```python
# Maybe we can show ourselves a bit...
plt.scatter(x="age", y="diameter", data=tree_data, c="condition")
```

```python
# And maybe, you can play around

scatter = plt.scatter(data=tree_data, x="age", y="diameter", c="condition")
plt.xlabel("Age (Years)")
plt.ylabel("Diameter (Cm)")
plt.title("Tree Condition by Age and Diameter")
handles, _ = scatter.legend_elements()
legend_labels = ["Excellent", "Good", "Poor"]
plt.legend(handles, legend_labels, title="Condition")
plt.show()
```

```python
# Plot 3: Box Plot
tree_data.boxplot(by="borough", column=["age"], figsize=(10,5)) #I am using base features because the syntax is simpler
```

```python
# Plot Let's-make-it-better-with-Seaborn: Box Plot
sns.boxplot(data=tree_data, x="borough", y="age")
```

```python
boxplot = sns.boxplot(data=tree_data, x="borough", y="age", palette=["#86373E", "#44332D", "#F1BD78"])
```

```python
# Plot-we-can-do-more-with-a-linear-model
linear = sns.lmplot(x="age", y="diameter", data=tree_data)
```

```python
# Plot-that-will-make-your-stats-professor-smile
linear = sns.lmplot(x="age",y="diameter", data=tree_data, line_kws={"color":"pink"})
```

```python
# On Wednesdays we wear pink

linear = sns.lmplot(x="age", y="diameter", data=tree_data, scatter_kws={"color": "#610C27"}, line_kws={"color": "pink"})
plt.xlabel("Age (Yrs)")
plt.ylabel("Diameter")
plt.title("Age vs Diameter of Trees")
plt.show()
plt.savefig("linear.pdf")# Step 5: Such Artwork should be eternized, so save your graphs!
linear = plt.savefig('linear.png', dpi=150)  # Save as PNG with higher resolution
```

```python
# Solution

#1 Import the libraries
import matplotlib.pyplot as plt

#2 Prepare the data
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
temperatures = [22, 24, 19, 23, 25, 27, 26]

#3 Create the bar chart
plt.figure(figsize=(10, 5))
plt.bar(days, temperatures, color='skyblue')

#4 Add titles and labels
plt.title('Daily Temperatures')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')

#5 Display the chart
plt.show()
```
